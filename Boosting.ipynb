```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Divorce_Y_N',axis=1)
y = df['Divorce_Y_N']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn import metrics
```

```python
RFC = RandomForestClassifier()
RFC.fit(X_train,y_train)
```

```python
forest_pred = RFC.predict(X_test)
```

```python
print(confusion_matrix(y_test,forest_pred))
print(classification_report(y_test,forest_pred))
metrics.accuracy_score(y_test,forest_pred)
```

```python
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(X_train,y_train)
```

```python
ada_pred = ada.predict(X_test)
```

```python
print(confusion_matrix(y_test,ada_pred))
print(classification_report(y_test,ada_pred))
metrics.accuracy_score(y_test,ada_pred)
```

```python
cm = confusion_matrix(y_test,ada_pred)
sns.heatmap(cm, annot=True,cmap="icefire")
plt.show()
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
gb_pred = gb.predict(X_test)
```

```python
print(confusion_matrix(y_test,gb_pred))
print(classification_report(y_test,gb_pred))
metrics.accuracy_score(y_test,gb_pred)
```

```python
cm2 = confusion_matrix(y_test,gb_pred)
sns.heatmap(cm2, annot=True,cmap="icefire")
plt.show()
```
